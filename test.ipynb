```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Load the dataset
data_dir = 'c:\\Users\\Asus\\Desktop\\DL-PotatoDisease\\training\\Potato'
img_size = (256, 256)
batch_size = 32

# Load the training dataset
train_data = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset='training',
    seed=200,
    image_size=img_size,
    batch_size=batch_size
)

# Get class names
class_names = train_data.class_names
print("Class names:", class_names)

# Display images with labels
def display_images_with_labels(dataset, class_names, num_images=10):
    plt.figure(figsize=(15, 15))
    count = 0
    for images, labels in dataset.take(1):
        for i in range(num_images):
            ax = plt.subplot(5, 5, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))
            label_index = labels[i].numpy()
            label = class_names[label_index]
            plt.title(f"{label} ({label_index})")
            plt.axis("off")
            count += 1
            if count >= num_images:
                break
    plt.show()

# Display the first 10 images with labels from the training dataset
display_images_with_labels(train_data, class_names)

# Additional debugging: Print out the labels for the first few images to verify
for images, labels in train_data.take(1):
    for i in range(10):
        label_index = labels[i].numpy()
        print(f"Image {i+1} - Label index: {label_index}, Class name: {class_names[label_index]}")
```
